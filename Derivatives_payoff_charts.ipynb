Using AMAZON CALL & PUT FOR ONE MONTH:

Strike Price :155

```python
op.single_plotter(spot=166.94, strike=155, op_type='c', tr_type='b', op_pr=14.25)
```

- Put Option:
Strike Price : 155
Spot Price as on 9th August, 2024: 166.94
Option Type: Put Option
Position : 1 Month
Premium : 1.46

```python
op.single_plotter(spot=166.94, strike=155, op_type='p', tr_type='b', op_pr=1.46)
```

#### STRATEGY:1 - BUY CALL & PUT OPTIONS SIMULTANEOUSLY

```python
op1={'op_type': 'p', 'strike': 155, 'tr_type': 'b', 'op_pr': 1.46} #Buy put option of Amazon
op2={'op_type': 'c', 'strike': 155, 'tr_type': 'b', 'op_pr': 14.25}#Buy call option Amazon

op_list=[op1, op2,]
op.multi_plotter(spot=166.94,spot_range=10, op_list=op_list)
```

#### STRATEGY:2 - SELL CALL & PUT OPTIONS SIMULTANEOUSLY

```python
op1={'op_type': 'p', 'strike': 155, 'tr_type': 's', 'op_pr': 1.46} #Sell put option of Amazon
op2={'op_type': 'c', 'strike': 155, 'tr_type': 's', 'op_pr': 14.25}#Sell call option Amazon

op_list=[op1, op2,]
op.multi_plotter(spot=166.94,spot_range=10, op_list=op_list)
```
